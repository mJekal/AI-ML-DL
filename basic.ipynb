```python
x=range(5)
y1=[i for i in range(1,6)]
y2=[i+100 for i in range(1,6)]

plt.figure()
plt.plot(x,y1)
plt.plot(x,y2)

plt.figure()
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
p1=ax1.plot(x,y1,'b')
p2=ax2.plot(x,y2,'r')
ax1.set_ylim([-10, 10])
ax1.set_xlabel("x")
ax1.set_ylabel("y1")
ax2.set_ylabel("y2")
ax1.set_title("y1 and y2")
ax1.grid()
plt.legend(p1+p2,["y1","y2"], loc="right")
```

```python
a = np.linspace(-1,1,100)
b = np.linspace(-1,1,100)
A,B = np.meshgrid(a,b)
L = A**2 + B**2
plt.contour(a,b,L,20)
plt.xlabel("a"); plt.ylabel("b"); plt.grid()
```

```python
import plotly.graph_objects as go

a=np.linspace(-2,2,100)
b=np.linspace(-2,2,100)
A, B = np.meshgrid(a,b)
L = np.cos(A**2.4)+np.sin(B**2.4)

fig = go.Figure(data=[go.Surface(x=a, y=b, z=L, colorscale="viridis", opacity=0.5)])
fig.update_traces(contours_z=dict(show=True, usecolormap=True, highlightcolor="limegreen", project_z=True))
fig.update_layout(title='cos(A^2.4)+sin(B^2.4)', width=700, height=600)
```

```python
plt.figure(figsize=(7,4))
for i in range(6):
  plt.subplot(2,3,i+1,xticks=[],yticks=[])
  plt.imshow(test_DS.data[i],cmap="gray")
  plt.title(test_DS.targets[i].item(),color="g")
```

```python
im=test_DS.data[0]
plt.figure(figsize=(20,15))
plt.imshow(im, cmap="gray")
plt.colorbar()

for i in range(im.shape[0]):
    for j in range(im.shape[1]):
        plt.text(j,i, im[i,j].item(),
                 horizontalalignment="center",
                 color="k" if im[i,j] > torch.max(im) / 1.5 else "w")
plt.xticks([])
plt.yticks([])
```
