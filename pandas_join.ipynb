```python
import pandas as pd
```

```python
d1 = {'id':[1,2,3,4,5,6],
      'col1':['O','O','O','T','T','T']}
df1 = pd.DataFrame(d1)
d2 = {'id':[2,4,6,8],
    'col2':['C','C','T','H']}
df2 = pd.DataFrame(d2)
```

```python
df1
```

```python
df2
```

### Inner 조인

```python
inner_join_result = pd.merge(df1, df2, on='id', how='inner')
inner_join_result
```

```python
inner_join_result = df1.merge(df2, on='id', how='inner')
inner_join_result
```

### left 조인

```python
left_outer_join_result = df1.merge(df2, on='id', how='left')
left_outer_join_result
```

```python
left_outer_join_result_01 = df2.merge(df1, on='id', how='left')
left_outer_join_result_01
```

### right 조인

```python
right_outer_join_result = df1.merge(df2, on='id', how='right')
right_outer_join_result
```

### outer 조인

```python
full_outer_join_result = pd.merge(df1, df2, on='id', how='outer')
full_outer_join_result
```
